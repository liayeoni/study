# WordCloud

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

# 앨리스 텍스트 파일 불러오기

```python
import pandas as pd
text = open('./06_alice.txt').read()
text
```

### 앨리스 이미지 파일 불러오기

```python
alice_mask = np.array(Image.open('./06_alice_mask.png'))
alice_mask #pixel
```

```python
stopwords = set(STOPWORDS) #별로 영향력없다고 판단되는 단어들
stopwords
```

```python
stopwords.add('said') #stopwords에 포함시켜라
```

### 앨리스 이미지 그리기

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

rc('font', family='Malgun Gothic')
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(alice_mask, cmap=plt.cm.gray)
```

### 단어별 빈도율 수치로 나타내기

```python
wc = WordCloud(
    background_color = 'white', max_words = 2000, mask=alice_mask, stopwords=stopwords
)

wc = wc.generate(text) # text데이터를 wordcloud로 생성
```

```python
wc.words_
```

### 그림 안에 표현

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

# StarWars 텍스트, 이미지 파일 불러오기

```python
text = open('./06_a_new_hope.txt').read()
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')

mask = np.array(Image.open('./06_stormtrooper_mask.png'))
```

```python
stopwords.add('int')
stopwords.add('ext')
```

```python
wc = WordCloud(
    max_words=1000, mask=mask, stopwords=stopwords, margin=10
).generate(text)
```

```python
default_colors = wc.to_array()
```

# grey색상으로 채우기

```python
import random

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs): # wordcloud의 guide에 맞게 함수 작성
    return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=13))
plt.axis('off')
plt.show()
```
