다음의 함수와 그 함수를 테스트하는 코드를 작성하세요. (PDF 파일 참고)

- 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수 (테스트 입력 : ‘1,000’)

- 주소를 입력받아 위도, 경도를 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)

제출 3.
- 각 함수 및 테스트 코드와 실행결과

```python
# 화폐단위 변환 함수
def won(money):
    return int(money.replace(',',''))

# Barnd-ID 반환 함수
def brand_id(brand):
    if brand == 'HD현대오일뱅크':
        brand = '현대오일뱅크'
        
    try:
        cursor.execute('select id from gas_brand where name = %s', (brand,))
        result = cursor.fetchall()

        return result[0][0]
    
    except:
        return 6

# 구 이름 반환 함수
def gu_name(address):
    gu_name = address.split(' ')[1]
    return gu_name

# 위도, 경도 반환 함수

def latlng(address):
    gmaps_key = 'gmaps_key'
    gmaps = googlemaps.Client(key=gmaps_key)
    tmp = gmaps.geocode(address, language = 'ko')
    if len(tmp) > 0:
        lat, lng = tmp[0]['geometry']['location']['lat'], tmp[0]['geometry']['location']['lng']
        return lat, lng
    return ''
```

```python
 # test won
result = won('1,000')
print(result, type(result))

# test id
result = brand_id('HD현대오일뱅크')
print(result)

# test gu_name
result = gu_name('서울시 강남구 헌릉로 730')
print(result)

# test latlng
result = latlng('서울시 강남구 헌릉로 730')
print(result)
```

```python
# 브랜드, 매장이름, 구, 전체 주소, 휘발유, 경유 가격, 셀프 주유, 세차, 충전, 경정비, 편의점, 24시간 영업 여부, 위도, 경도 정보

def insert_info():
    req = driver.page_source #주유소 정보 페이지를 BeatifulSoup으로 처리
    soup = BeautifulSoup(req, 'html.parser')

    #브랜드 이름
    brand_info = soup.select('#poll_div_nm')
    brand = brand_info[0].text
    brand

    #brand
    id_info = brand_id(brand)
    id_info

    #매장 이름
    name_info = soup.select('#os_nm')
    name = name_info[0].text

    #주소
    address_info = soup.select('#rd_addr')
    address = address_info[0].text
    address

    #구
    gu = gu_name(address)
    gu

    #휘발유 가격
    gas_info = soup.select('#b027_p')
    gas_price = won(gas_info[0].text)
    gas_price

    #경유 가격
    diesel_info = soup.select('#d047_p')
    diesel_price = won(diesel_info[0].text)
    diesel_price

    # 셀프 주유
    self_yn = soup.select('#self_icon')
    if len(self_yn) > 0:
        self_yn = 1
    else:
        self_yn = 0
    self_yn

    #세차장
    wash_yn = soup.select('#cwsh_yn')
    if wash_yn[0].get('src') == '/images/user/gis/oil_station_service1_01.gif':
        wash_yn = 1
    else:
        wash_yn = 0

    #충전소
    lpg_yn = soup.select('#lpg_yn')
    if lpg_yn[0].get('src') == '/images/user/gis/oil_station_service1_02_01.gif':
        lpg_yn = 1
    else:
        lpg_yn = 0

    #경정비
    maint_yn = soup.select('#maint_yn')
    if maint_yn[0].get('src') == '/images/user/gis/oil_station_service1_03.gif':
        maint_yn = 1
    else:
        maint_yn = 0

    #편의점
    cvs_yn = soup.select('#cvs_yn')
    if cvs_yn[0].get('src') == '/images/user/gis/oil_station_service1_04.gif':
        cvs_yn = 1
    else:
        cvs_yn = 0

    #편의점
    sel24_yn = soup.select('#sel24_yn')
    if sel24_yn[0].get('src') == '/images/user/gis/oil_station_service1_05.gif':
        sel24_yn = 1
    else:
        sel24_yn = 0

    #위도, 경보
    lat = latlng(address)[0]
    lng = latlng(address)[1]
    
    #sql insert 쿼리문
    data_list = (id_info, name, gu, address, gas_price, diesel_price, wash_yn, self_yn, lpg_yn, maint_yn, cvs_yn, sel24_yn, lat, lng )
    sql = 'insert into gas_station values(NULL, %s, %s, "서울", %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)' 
    cursor.execute(sql, data_list)
    result = cursor.fetchall()
    print('OK')
    conn.commit()
```

문제 6.

위도, 경도 정보를 이용하여 미왕빌딩에서 1킬로 이내에 위치한 주유소 정보를 검색하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 미왕빌딩으로 부터의 거리 (km)

제출 6.
- 쿼리, 결과

```python
#미왕빌딩 위도, 경도 : 37.49570, 127.0294
sql = "select s.id, b.name as brand, s.name, s.address, (6371*acos(cos(radians(37.49570))*cos(radians(s.lat))*cos(radians(s.lng)-radians(127.0294))+sin(radians(37.49570))*sin(radians(s.lat)))) as distance " + \
       "from GAS_BRAND as b, GAS_STATION as s " + \
       "where s.brand = b.id " + \
       "having distance <= 1 " + \
       "order by distance"

cursor.execute(sql)
result = cursor.fetchall()

for i in result:
    print(i)
```

문제 7.

위도, 경도 정보를 이용하여 미왕빌딩에서 셀프주유가 가능하고 24시간이면서 편의점이 있는 가장 가까운 주유소 10개를 휘 발유 가격이 가장 저렴한 순으로 정렬하여 조회하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 휘발유 가격, 부가정보 (셀프, 24시간, 편의점 여부), 미왕빌딩으로 부터의 거리 (km)


제출 7.
- 쿼리, 결과

```python
sql = "select * from (select s.id, b.name as brand, s.name, s.address, gasoline, self, 24_hours, convenience_store, (6371*acos(cos(radians(37.49570))*cos(radians(s.lat))*cos(radians(s.lng)-radians(127.0294))+sin(radians(37.49570))*sin(radians(s.lat)))) as distance " + \
       "from GAS_BRAND as b, GAS_STATION as s " + \
       "where s.brand = b.id " + \
       "and self = 1 and 24_hours = 1 and convenience_store = 1 " + \
       "order by distance limit 10) as tmp " + \
       "order by tmp.gasoline"

cursor.execute(sql)
result = cursor.fetchall()
for i in result:
    print(i)
```

제출 8.

```python
sql = 'select s.gu, b.name, avg(s.gasoline) as avg_gasoline from gas_station s, gas_brand b where b.id = s.brand group by gu, b.name order by avg(gasoline)'
cursor.execute(sql)
result = cursor.fetchall()
for i in result:
    print(i)
```
